```python
import tensorflow as tf
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
import numpy as np

df=pd.read_excel(r"C:\Users\weije\Desktop\Kaggle Data\sales_data_sample.xlsx")

X=df["QUANTITYORDERED"].values.reshape(-1,1)
y=df["SALES"]
plt.scatter(X,y)

print(df.columns)
year_id_Quantity_per_year = df.groupby("YEAR_ID").PRICEEACH.mean().reset_index()

regr= LinearRegression()
regr.fit(X,y)
print(regr.coef_[0])
y_predict = regr.predict(X)
plt.plot(X, y_predict, color='red')
plt.show()
X_future=np.array(range(2003,2005))
X_future=X_future.reshape(-1,1)
future_predict=regr.predict(X_future)
plt.plot(future_predict,X_future)
```
