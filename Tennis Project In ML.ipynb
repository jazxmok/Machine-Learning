```python
import seaborn as sb
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression

df=pd.read_csv(r"C:\Users\weije\Downloads\tennis_stats.csv")

print(df.columns)

X = df[["BreakPointsOpportunities"]]
y = df['Ranking']

plt.scatter(X,y)

features = df[['FirstServeReturnPointsWon']]
outcome = df[['Winnings']]

features_train, features_test, outcome_train, outcome_test = train_test_split(features, outcome, train_size = 0.8)
model = LinearRegression()
model.fit(features_train,outcome_train)
model.score(features_test,outcome_test)
prediction = model.predict(features_test)
plt.scatter(outcome_test,prediction, alpha=0.4)
model_xx = model.score(features, outcome)
print('Overall R-squared:', model_xx)

Breakeven=df["BreakPointsOpportunities"].values.reshape(-1,1)
Breakeven, Breakeven_test, y_train, y_test = train_test_split(Breakeven, y, test_size=0.2, random_state=0)
model = LinearRegression()
model.fit(Breakeven_train, y_train)

score = model.score(Breakeven_test, y_test)
print('R-squared:', score)
features = df[['BreakPointsOpportunities',
'FirstServeReturnPointsWon']]
outcome = df[['Winnings']]
```
